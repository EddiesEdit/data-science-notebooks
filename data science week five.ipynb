## Handling Outliers

```python
sns.boxplot(x=df['Fare'])
plt.show()
```

```python
Q1 = df['Fare'].quantile(0.25)
Q3 = df['Fare'].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

outliers = df[(df['Fare'] < lower) | (df['Fare'] > upper)]
outliers.shape
```

```python
outliers
```

```python
df
```

```python
df['Fare'] = np.clip(df['Fare'], lower, upper)
```
