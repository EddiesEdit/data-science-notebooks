# MODULE 5: Inferential Statistics 


# 1. Introduction to Inferential Statistics

## What is Inferential Statistics?

Inferential statistics is the branch of statistics that helps us:

* Make conclusions about a population using a sample
* Make predictions or generalizations
* Test ideas (hypotheses)

### Simple idea:

We don’t study everyone (population). We study a **sample** and infer about the whole.

### Example:

* Population: All students in a university
* Sample: 100 students
* Question: Do students study more than 5 hours daily?

We use inferential statistics to answer that.

---

# 2. Data Distributions

## What is a distribution?

A distribution shows how data values are spread.

### Common distributions:

## (1) Normal Distribution

* Bell-shaped curve
* Symmetrical
* Mean = Median = Mode

Example: Heights, IQ scores

## (2) Skewed Distribution

* Left skewed (tail on left)
* Right skewed (tail on right)

## (3) Uniform Distribution

* All values have equal probability

---

# 3. The 6 Must-Know Statistical Distributions

## 1. Normal Distribution

Used in: heights, exam scores

## 2. Binomial Distribution

Used when:

* Only 2 outcomes (success/failure)
  Example: coin toss, pass/fail

## 3. Poisson Distribution

Used for:

* Counting events over time
  Example: number of calls per hour

## 4. Uniform Distribution

All outcomes are equally likely

## 5. Exponential Distribution

Models time between events
Example: time between bus arrivals

## 6. t-Distribution

Used when sample size is small

---

# 4. Central Limit Theorem (CLT)

## Definition

The CLT says:

> If you take many samples from any distribution, the sample means will form a normal distribution.

Even if original data is NOT normal.

## Why it matters

* It allows us to use normal distribution in real-world problems
* Foundation of hypothesis testing

## Key idea:

As sample size increases:

* Distribution becomes more normal

---

# 5. Sampling, Estimation & CLT

## Sampling

Selecting a subset from a population.

### Types:

* Random sampling
* Stratified sampling
* Systematic sampling

## Estimation

Using sample data to estimate population parameters.

### Types:

* Point estimation (single value)
* Interval estimation (range)

Example:

* Mean salary = $50,000 ± $2,000

---

# 6. Hypothesis Testing

## What is a hypothesis?

A claim or assumption we want to test.

### Example:

* H0: Students study 5 hours/day
* H1: Students do NOT study 5 hours/day

## Null Hypothesis (H0)

* Assumes no effect
* Default assumption

## Alternative Hypothesis (H1)

* What we want to prove

---

## Steps in Hypothesis Testing

1. State H0 and H1
2. Choose significance level (α = 0.05)
3. Compute test statistic
4. Compare p-value
5. Accept or reject H0

---

# 7. Test Statistics

## What is a test statistic?

A value calculated from sample data used to test hypotheses.

### Common ones:

* z-score
* t-score
* chi-square
* F-statistic

---

# 8. Chi-Square Test

## Used for:

* Categorical data
* Relationship between variables

### Example:

Is gender related to product preference?

## Formula idea:

Chi-square compares:

* Observed values
* Expected values

---

# 9. ANOVA (Analysis of Variance)

## Used for:

Comparing means of 3 or more groups

### Example:

Do 3 teaching methods affect student performance differently?

## Key idea:

* Compares variance within groups vs between groups

---

# 10. Summary of Statistical Tests

| Test       | Purpose                   |
| ---------- | ------------------------- |
| t-test     | Compare 2 means           |
| z-test     | Large sample mean test    |
| Chi-square | Categorical relationships |
| ANOVA      | Compare 3+ means          |
