### EXCEPTION HANDLING
Errors = inevitable
because they know errors are inevitable **EXCEPTION HANDLING** is created to avoid errors that might cause problems for your software or hardware

```python
# Example 
try:
    x = int("abc")
except:
    print("Error occurred")
```

```python
x= int("abc")
```

```python
#Specific Exceptions very good for trobleshooting or debegging
try:
    num = int(input("Enter number: "))
    print(10 / num)
except ZeroDivisionError:
    print("Cannot divide by zero")
except ValueError:
    print("Invalid input")
```

```python
# Custom Exception

def withdraw(amount):
    if amount > 1000:
        raise Exception("Insufficient funds")

try:
    withdraw(2000)
except Exception as e:
    print(e)
```

### REGULAR EXPRESSIONS
Most Useful Regular Expressions (Regex)
|  Expression |  Meaning                        |  Example        |  Matches          |
| ------------- | --------------------------------- | ----------------- | ------------------ |
| `\d`          | Any digit (0–9)                   | `"Age: 25"`       | `2, 5`             |
| `\d+`         | One or more digits                | `"123 apples"`    | `123`              |
| `\D`          | Non-digit                         | `"A1B"`           | `A, B`             |
| `\w`          | Word character (a-z, A-Z, 0-9, _) | `"user_1"`        | `u, s, e, r, _, 1` |
| `\w+`         | One or more word characters       | `"user_123"`      | `user_123`         |
| `\W`          | Non-word character                | `"hello!"`        | `!`                |
| `\s`          | Whitespace (space, tab)           | `"Hello World"`   | space              |
| `\s+`         | One or more spaces                | `"Hello   World"` | spaces             |
| `\S`          | Non-whitespace                    | `"Hi there"`      | `H, i, t, h...`    |
| `\S+`         | Continuous non-space text         | `"hello world"`   | `hello`, `world`   |



Common Real-World Patterns

|  Expression |  Meaning             |  Example         |  Matches        |
| ------------- | ---------------------- | ------------------ | ---------------- |
| `r"\S+@\S+"`  | Simple email pattern   | `"test@gmail.com"` | `test@gmail.com` |
| `r"\d{3}"`    | Exactly 3 digits       | `"1234"`           | `123`            |
| `r"\d{2,4}"`  | Between 2 and 4 digits | `"2025"`           | `2025`           |
| `r"[a-z]"`    | Lowercase letters      | `"abc"`            | `a, b, c`        |
| `r"[A-Z]"`    | Uppercase letters      | `"ABC"`            | `A, B, C`        |
| `r"[a-zA-Z]"` | Any letter             | `"Hello"`          | `H, e, l...`     |
| `r"[0-9]"`    | Digits (same as `\d`)  | `"456"`            | `4, 5, 6`        |




```python
# Example 
import re

text = "Contact: user123@gmail.com, Age: 25"

# Extract email
email = re.findall(r"\S+@\S+", text)

# Extract numbers
numbers = re.findall(r"\d+", text)

print("Email:", email)
print("Numbers:", numbers)
```
